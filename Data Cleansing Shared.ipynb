# 1. INTRODUCTION

**📌 Definisi Data Cleansing**

Data cleansing (atau data cleaning / data scrubbing) adalah proses membersihkan, memperbaiki, dan menstandarkan data agar bebas dari kesalahan, inkonsistensi, maupun informasi yang tidak relevan.

Tujuan utamanya adalah memastikan data akurat, konsisten, dan siap digunakan untuk analisis atau pengambilan keputusan.

**📌 Pentingnya Data Cleansing dalam Analisis**

- Kualitas data menentukan kualitas analisis.
Garbage in = garbage out → jika data yang masuk kotor, maka hasil analisis pasti menyesatkan.

- Mengurangi risiko kesalahan keputusan. Data yang salah bisa membuat manajemen salah strategi.

- Meningkatkan efisiensi. Data bersih lebih mudah diproses, mengurangi waktu untuk troubleshooting.

- Membuat model analisis/prediksi lebih akurat. Data kotor bisa menyebabkan bias atau error yang tinggi.

**📌 Dampak Data Kotor terhadap Hasil Analisis & Keputusan Bisnis**

1. Kesalahan Perhitungan

- Contoh: Ada transaksi ganda (duplikat), revenue perusahaan terlihat lebih besar dari kenyataan.

2. Insight yang Menyesatkan

- Contoh: Data umur ada yang negatif → saat analisis demografi pelanggan, hasil menjadi tidak valid.

3. Keputusan Salah

- Contoh: Karena ada data missing di cabang tertentu, perusahaan mengira cabang tersebut merugi → padahal datanya hanya tidak lengkap.

4. Meningkatkan Biaya & Risiko

- Data pelanggan yang salah alamat → produk tidak sampai, biaya operasional meningkat, kepuasan pelanggan turun.

## 2.1 Missing Data

**🔹 Apa itu Missing Data?**

Missing data adalah kondisi di mana suatu nilai dalam dataset tidak tersedia atau kosong.
Bisa ditandai dengan:

- Sel kosong (blank)

- Nilai NaN / NULL

- Tanda khusus seperti -, ?, atau "unknown"

**🔹 Penyebab Missing Data**

1. Human error → salah input, lupa isi kolom.

2. Kesalahan sistem → bug aplikasi, data tidak tersimpan.

3. Respon tidak lengkap → survei/kuisioner tidak dijawab semua.

4. Tidak relevan → misalnya kolom "Tanggal Pernikahan" untuk orang yang belum menikah.

**🔹 Dampak Missing Data**

1. Statistik/analisis bisa bias.

2. Machine learning model bisa gagal belajar (karena banyak algoritma tidak menerima nilai kosong).

3. Bisa menyebabkan keputusan bisnis salah (misalnya jumlah pelanggan aktif dihitung kurang).

**🔹 sytax**

| Tujuan                                | Syntax Python (pandas)                                                  |
| ------------------------------------- | ----------------------------------------------------------------------- |
| **Cek missing value (NaN/null)**      | `df.isnull()` atau `df.isna()`                                          |
| **Hitung total missing per kolom**    | `df.isnull().sum()`                                                     |
| **Hapus baris dengan missing value**  | `df.dropna()`                                                           |
| **Hapus kolom dengan missing value**  | `df.dropna(axis=1)`                                                     |
| **Isi missing dengan nilai tertentu** | `df.fillna(0)` atau `df.fillna("Unknown")`                              |
| **Isi missing dengan mean/median**    | `df["kolom"].fillna(df["kolom"].mean())`                                |
| **Isi missing per grup/kategori**     | `df["kolom"].fillna(df.groupby("kategori")["kolom"].transform("mean"))` |
| **Tambahkan flag kolom penanda**      | `df["kolom_missing"] = df["kolom"].isnull().astype(int)`                |

## 2.2 Duplicate Data

**🔹 Apa itu Duplicate Data?**

Duplicate data adalah kondisi di mana ada baris (record) yang sama persis atau sangat mirip dalam dataset, sehingga terjadi pengulangan informasi.

**🔹 Penyebab Duplicate Data**

1. Input ganda oleh pengguna (double entry).

2. Proses impor data berulang dari sistem yang sama.

3. Penggabungan dataset tanpa cek keunikan ID.

4. Kesalahan sistem yang merekam transaksi lebih dari sekali.

**🔹 Dampak Duplicate Data**

1. Analisis bias → misalnya jumlah pelanggan dihitung dobel.

2. Keputusan salah → pelanggan dianggap lebih banyak dari kenyataan.

3. Boros penyimpanan → data tersimpan berulang tanpa manfaat.

4. Kesulitan integrasi → data ID yang sama muncul berkali-kali.

**🔹 sytax**

| Tujuan                                     | Syntax Python (pandas)                 |
| ------------------------------------------ | -------------------------------------- |
| **Cek duplikat**                           | `df.duplicated()`                      |
| **Hitung jumlah duplikat**                 | `df.duplicated().sum()`                |
| **Hapus duplikat (simpan pertama)**        | `df.drop_duplicates()`                 |
| **Hapus duplikat (simpan terakhir)**       | `df.drop_duplicates(keep="last")`      |
| **Hapus duplikat berdasar kolom tertentu** | `df.drop_duplicates(subset=["kolom"])` |
| **Tandai baris duplikat**                  | `df["is_duplicate"] = df.duplicated()` |

## 2.3 Outlier

**🔹 Apa itu Outlier?**

Outlier adalah data yang nilainya jauh berbeda / ekstrem dibanding sebagian besar data lain dalam dataset.

Bisa muncul karena error input, anomali nyata, atau variasi alami.

**🔹 Dampak Outlier**

1. Menggeser rata-rata (mean) → hasil analisis jadi bias.

2. Membuat grafik/visualisasi jadi tidak representatif.

3. Bisa menyesatkan model machine learning (prediksi jadi buruk).

**🔹 Cara Deteksi Outlier**

```python
from IPython.display import Image
Image(filename='Image/Data Cleansing outliers detect.jpg')
```

**🔹 Cara Menangani Outlier**

1. Hapus outlier → bila jelas data error (misalnya umur = -10).

2. Transformasi data → log transform, square root, dsb.

3. Winsorizing → batasi nilai ekstrem ke batas tertentu (misalnya semua nilai di atas P95 diganti P95).
4. Pisahkan data outlier → buat analisis khusus (misalnya high spender di e-commerce).

**🔹 Syntax**

| Tujuan                           | Syntax Python (pandas)                                                                                        |                              |
| -------------------------------- | ------------------------------------------------------------------------------------------------------------- | ---------------------------- |
| Cek distribusi kolom numerik     | `df["kolom"].describe()`                                                                                      |                              |
| Deteksi outlier dengan Z-score   | `from scipy import stats; z = stats.zscore(df["kolom"])`                                                      |                              |
| Deteksi outlier dengan IQR       | `Q1 = df["kolom"].quantile(0.25); Q3 = df["kolom"].quantile(0.75); IQR = Q3-Q1; df[(df["kolom"] < Q1-1.5*IQR) | (df["kolom"] > Q3+1.5*IQR)]` |
| Hapus outlier (misalnya via IQR) | `df = df[(df["kolom"] >= Q1-1.5*IQR) & (df["kolom"] <= Q3+1.5*IQR)]`                                          |                              |
| Log transform                    | `df["kolom_log"] = np.log(df["kolom"])`                                                                       |                              |

**iii. Standart Deviations**

Standar deviasi adalah ukuran yang menunjukkan seberapa besar penyebaran suatu kumpulan data terhadap nilai rata-ratanya.

Nilai yang berada setidaknya 3 standar deviasi dari rata-rata dianggap sebagai outlier (data pencilan).
- Metode ini digunakan untuk data yang berdistribusi normal, atau berbentuk seperti kurva lonceng (bell-shaped).
- Batas 3 standar deviasi ini dapat diubah menjadi 2 atau lebih dari 4, tergantung pada karakteristik data yang dianalisis.

```python
from IPython.display import Image
Image(filename='Image/Data Cleansing std.jpg')
```

```python
import numpy as np

# calculate the mean and standard deviation
mean = np.mean(df.Grade)
sd = np.std(df.Grade)

# show mean and standard deviation
mean, sd
```

```python
# identify points that are more than 3 standard deviations away
[grade for grade in df.Grade if (grade < mean - 3*sd) or (grade > mean + 3*sd)]
```

Batas bawah (lower limit)
= mean − 3 × sd
= 84 − 3 × 8.6
= 84 − 25.8
= 58.2

Batas atas (upper limit)
= mean + 3 × sd
= 84 + 25.8
= 109.8

```python
# identify points that are more than 2 standard deviations away
[grade for grade in df.Grade if (grade < mean - 2*sd) or (grade > mean + 2*sd)]
```

**🎯 1️⃣ Konsep dasarnya**

Dalam distribusi normal (bell curve):

| Batas SD dari mean | Cakupan data (±) | Data di luar batas (outlier) |
| ------------------ | ---------------- | ---------------------------- |
| ±1 SD              | ~68%             | ~32%                         |
| ±2 SD              | ~95%             | ~5%                          |
| ±3 SD              | ~99.7%           | ~0.3%                        |
| ±4 SD              | ~99.99%          | ~0.01%                       |



Artinya:

- ±2 SD → sebagian kecil data bisa dianggap outlier

- ±3 SD → hanya data ekstrem banget yang dianggap outlier

- ±4 SD → hampir semua data dianggap normal, kecuali yang benar-benar aneh

**⚙️ 2️⃣ Kapan pakai 2 SD**

Gunakan 2 SD jika:

- Kamu ingin mendeteksi outlier secara sensitif (lebih banyak nilai akan terdeteksi).

- Data tidak terlalu besar dan kamu ingin menangkap variasi lebih kecil.

Contoh kasus:

Nilai ujian siswa — ingin tahu siapa yang nilainya jauh di bawah rata-rata.

Data kualitas produksi — ingin cepat tahu produk yang sedikit keluar standar.

🧠 Biasanya cocok untuk data operasional atau kualitas (QC, KPI, performa karyawan).

**⚙️ 3️⃣ Kapan pakai 3 SD**

Gunakan 3 SD jika:

- Kamu ingin hanya mendeteksi outlier ekstrem.

- Dataset kamu besar dan stabil, sehingga outlier 2 SD terlalu banyak muncul.

- Kamu mengikuti aturan standar statistik (empirical rule).

📊 Contoh:

Analisis data ilmiah / eksperimen laboratorium

Data keuangan atau hasil pengukuran yang stabil

Deteksi error input yang jelas salah

🧠 Ini adalah standar paling umum — dipakai di banyak konteks statistik (termasuk Six Sigma).

**⚙️ 4️⃣ Kapan pakai 4 SD**

Gunakan 4 SD jika:

- Kamu hanya ingin menghapus outlier ekstrem sekali (jarang terjadi).

- Data kamu memiliki variasi alami besar atau distribusi tidak normal (skewed).

- Tujuanmu adalah tidak menghapus data valid yang ekstrem tapi masih wajar.

📊 Contoh:

Data pendapatan, lama kerja, nilai saham — di mana ada ekor panjang (long tail).

Analisis data manusia atau bisnis yang sangat variatif.

| Tujuan Analisis                                | SD yang Disarankan                        |
| ---------------------------------------------- | ----------------------------------------- |
| Menemukan semua nilai mencurigakan             | ±2 SD                                     |
| Menemukan hanya outlier ekstrem                | ±3 SD                                     |
| Data sangat bervariasi (hindari false outlier) | ±4 SD                                     |
| Analisis eksploratif (awal)                    | Mulai dari 2 SD, lalu bandingkan hasilnya |

### 2.3.2 Handle Outliers

Like with missing data, there are multiple ways to handle outliers:
- Keep outliers
- Remove an entire row or column with outliers
- Impute outliers with NaN or a substitute like the average, mode, max, etc.
- Resolve outliers based on your domain expertise

#### i. Remove an entire row or column

```python
# these are the two rows with outliers
df[df.Grade < 60]
```

```python
# option 1: drop the two rows with outliers
df.drop([36, 53])
```

```python
# option 2: this is the opposite of the result above
df[df.Grade >= 60]
```

#### ii. Impute the data

```python
# identify the outliers
df.Grade < 60
```

```python
# locate the outliers
df[df.Grade < 60]
```

```python
# minimum of the non-outliers
min_grade = df[df.Grade > 60].Grade.min()
min_grade
```

```python
# update the outlier values with the minimum grade
df.Grade = np.where(df.Grade < 60, min_grade, df.Grade)
df.Grade
```

```python
# confirm the updated values
df.loc[[37, 56]]
```

#### iii. Resolve outliers based on domain expertise

```python
# see John's information
df[df.Student == 'John']
```

```python
# his last test wasn't logged and he should have an 74
df.loc[37, 'Grade'] = 74
```

```python
# confirm the updated value
df[df.Student == 'John']
```

# LABEL ENCODING

### 📌 Categorical Data Encoding

**🔹 Apa itu Categorical Data?**

Categorical data adalah data yang berupa kategori atau label, bukan angka.

- Nominal: tidak punya urutan (contoh: jenis kelamin, kota, warna).

- Ordinal: punya urutan/ranking (contoh: pendidikan [SMA < S1 < S2 < S3]).

👉 Banyak algoritma machine learning tidak bisa langsung membaca teks, sehingga perlu diubah menjadi angka → proses ini disebut encoding.

**🔹 Jenis Encoding**

**1. Label Encoding**

        - Mengubah setiap kategori menjadi angka urut.
        - Contoh
    
| Gender | Label |
| ------ | ----- |
| Male   | 0     |
| Female | 1     |

        - Kelemahan: ada kesan urutan, padahal kategori nominal.

        - Syntax

*from sklearn.preprocessing import LabelEncoder*

*le = LabelEncoder()*

*le.fit_transform(df["Class"])*

```python
df_copy = df.copy()
df_copy
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_copy['Class_Label'] = le.fit_transform(df["Class"])
df_copy
```

```python
df_copy.groupby('Class')['Class_Label'].unique()
```

**🔹 Jenis Encoding**

**2. One-Hot Encoding**

        - Membuat kolom baru untuk tiap kategori dengan nilai 0/1.
        - Contoh
    
| Kota    | Kota_Jakarta | Kota_Bandung | Kota_Surabaya |
| ------- | ------------ | ------------ | ------------- |
| Jakarta | 1            | 0            | 0             |
| Bandung | 0            | 1            | 0             |

        - Cocok untuk nominal (tidak ada urutan).

        - Syntax

*pd.get_dummies(df["Kota"], prefix="Kota")*

```python
pd.get_dummies(df_copy["Class"], prefix="Class")
```

**🔹 Jenis Encoding**

**3. Ordinal Encoding**

        - Memberi angka berdasarkan urutan logis kategori..
        - Contoh (Pendidikan):
    
| Pendidikan | Ordinal |
| ---------- | ------- |
| SMA        | 1       |
| S1         | 2       |
| S2         | 3       |
| S3         | 4       |


        - Cocok untuk ordinal data.

        - Syntax

*from sklearn.preprocessing import OrdinalEncoder*

*ord_enc = OrdinalEncoder(categories=[['Freshman','Sophomore','Junior','Senior']])*

*df["Year_Ordinal"] = ord_enc.fit_transform(df[["Year"]])*


```python
df_copy['Year'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder(categories=[['Freshman','Sophomore','Junior','Senior']])

df_copy["Year_Ordinal"] = ord_enc.fit_transform(df_copy[["Year"]])
df_copy
```

**🔹 Jenis Encoding**

**Frequency / Count Encoding**

        - Mengganti kategori dengan frekuensi kemunculannya.
        - Contoh (Pendidikan):
    
| Kota     | Frequency |
| -------- | --------- |
| Jakarta  | 100       |
| Bandung  | 80        |
| Surabaya | 50        |

        - Syntax
        
*freq_map = df["Kota"].value_counts().to_dict()*

*df["Kota_Freq"] = df["Kota"].map(freq_map)*


```python
freq_map = df_copy["Class"].value_counts().to_dict()
df_copy["Classa_Freq"] = df_copy["Class"].map(freq_map)
df_copy
```

**🔹 Rangkuman: Kapan Gunakan Encoding?**

- Label Encoding: kategori biner (Male/Female).

- One-Hot Encoding: kategori nominal tanpa urutan.

- Ordinal Encoding: kategori dengan urutan logis.

- Frequency Encoding: kategori dengan banyak level (misalnya ratusan kota).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Contoh data ---
np.random.seed(42)
df = pd.DataFrame({
    'Umur': np.random.normal(30, 5, 100).tolist() + [120, 150],  # ada outlier
    'Gaji': np.random.normal(7000, 1500, 100).tolist() + [50000, 60000],  # ada outlier
})

# --- Fungsi untuk menangani outlier dengan IQR ---
def handle_outlier_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return np.where(series < lower, lower,
           np.where(series > upper, upper, series))

# --- Loop tiap kolom numerik ---
for col in df.select_dtypes(include=np.number).columns:
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    # Plot sebelum handling
    sns.boxplot(y=df[col], ax=axes[0], color="salmon")
    axes[0].set_title(f"Before Handling Outlier - {col}")

    # Handling outlier
    df[f"{col}_handled"] = handle_outlier_iqr(df[col])

    # Plot sesudah handling
    sns.boxplot(y=df[f"{col}_handled"], ax=axes[1], color="lightgreen")
    axes[1].set_title(f"After Handling Outlier - {col}")

    plt.tight_layout()
    plt.show()
```
